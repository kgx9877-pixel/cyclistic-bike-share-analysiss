```python
import pandas as pd

df1 = pd.read_csv("202004-divvy-tripdata.csv")
df2 = pd.read_csv("202005-divvy-tripdata.csv")
df3 = pd.read_csv("202006-divvy-tripdata.csv")
df = pd.concat([df1, df2, df3])
df.head()
df.dropna(inplace=True)
df['ride_length'] = pd.to_datetime(df['ended_at']) - pd.to_datetime(df['started_at'])
df['member_casual'].value_counts()
```

```python
df['ride_length'].mean()
```

```python
df.groupby('member_casual')['ride_length'].mean()
```

```python
df['day'] = pd.to_datetime(df['started_at']).dt.day_name()
df['day'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['member_casual'].value_counts().plot(kind='bar')
plt.show()
```

## Insights

1. Casual riders have a higher average ride duration compared to members.
   This suggests that casual users mainly use bikes for leisure and recreational purposes.

2. Members take rides more frequently.
   This indicates that they primarily use the service for daily commuting such as work or routine travel.

3. Ride activity is higher on weekends (Saturday and Sunday).
   This shows that casual users prefer riding during their free time.

## Conclusion

To convert casual riders into annual members, the company should:

- Offer special discounts and promotions on weekends
- Highlight membership benefits for longer rides
- Promote membership plans through the app and advertisements
- Encourage frequent riders with subscription-based offers
